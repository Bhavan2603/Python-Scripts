22. Visualize the distribution of BasePay, OvertimePay, and OtherPay using histograms or kernel density plots

```python
col = df[['BasePay', 'OvertimePay', 'OtherPay']]

for i in col:
    
    plt.figure(figsize=(10, 5))
    plt.subplot(1, 2, 1)
    plt.hist(col[i], bins=10, color='red', edgecolor='black')
    plt.xlabel(i)
    plt.xticks(rotation=90)
    plt.ylabel('Frequency')
    plt.title('Distribution of ' + i)
    
    plt.subplot(1, 2, 2)
    df[i].plot(kind='kde', color='green')
    plt.xlabel(i)
    plt.xticks(rotation=90)
    plt.ylabel('Density')
    plt.title('Distribution of ' + i)
    
    plt.show()
```
