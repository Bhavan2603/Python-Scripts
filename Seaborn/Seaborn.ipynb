### **Data Visualization using Seaborn**
#### **1. Bar Plot/ Count Plot**
- Displaying the distribution of categorical variables.
- Visualizing the frequency or count of each category in a dataset.

```python
df.groupby('day')['tip'].sum()
```

```python
ac = sns.barplot(data=df, x='day', y='tip', estimator='mean', errorbar=None)
ac.bar_label(ac.containers[0], fontsize=10)
ac.plot('Sun', 3.5, 'r*', markersize=10)
plt.grid()
plt.xlabel('Days')
plt.ylabel('Tips')
plt.title('Bar Plot of Tips by Days')
plt.show()
```

#### **2. Count Plot**
- Shows the count of observations in each categorical bin using bars.

```python
sns.countplot(data=df, x='day', hue='smoker', palette='Greens')
plt.grid(True)
plt.title('Count of Tips by Days')
plt.show()
```

#### **3. Pie Chart**

```python
Smoker = df['smoker'].value_counts()
Smoker
```

```python
plt.pie(Smoker, labels=Smoker.index, autopct='%1.f%%', explode=[0, 0.1], colors=sns.color_palette('Set1'))
plt.legend()
plt.title('Distribution of Smoker')
plt.show()
```

```python
sns.color_palette('summer')
```

#### **4. Scatter Plot**
- Exploring the relationship between two continuous variables.
- Identifying patterns, correlations, or in the data.

```python
df['size'].unique()
```

```python
sns.scatterplot(data=df, x='total_bill', y='tip', alpha=0.5, marker='*', edgecolor='black', legend=True, size='size', sizes=(20, 200))
plt.title('Scatter Plot of Total Bill by Tips')
plt.show()
```

#### **5. Line Plot**
- Displaying the trend or pattern in a time series.
- Showing the relationship between two continuous variables over a continuous inerval.
- Comparing changes in variables over a continuous range.

```python
sns.lineplot(data=df, x='day', y='tip', hue='sex', orient='x', marker='*', markersize=10, linestyle=':')
plt.title('Line Plot of Tip over Total_bill')
plt.show()
```

#### **6. Box Plot**
- Displaying the mean, median, quantiles, and outliers in the data.
- Comparing the distribution of multiple variables.
- Identifying the spread of numerical variables.
- Detecting potential outliers in the dataset.

```python
sns.boxplot(data=df, y='total_bill', fill=True, palette='Set3')
plt.title('Box Plot of Total Bill')
plt.show()
```

#### **7. Density Plot**
- Visualizing the distribution of a continuous variable.
- Identifying the peaks, valeeys, and overall patterns in the data.
- Understanding the shape of the distribution.
- Comparing the distributions of multiple variables.

```python
sns.distplot(df['total_bill'], hist=True, bins=20, color='b', kde=True, vertical=False)
plt.title("Distribution Plot of Total Bill")
plt.show()
```

```python
sns.kdeplot(data=df, x='total_bill', hue='time', shade=True)
plt.title('Kernel Density Estimation Plot of Total Bill')
plt.show()
```

```python
sns.histplot(data=df, x='total_bill', edgecolor='black', bins=10, element='step', color='blue')
plt.title('Hist Plot of Total Bill')
plt.show()
```

#### **8. Pair Plot**
Exploring correlations and trends between multiple variables by visualizing them in pairs.

```python
sns.pairplot(df, kind='scatter')
#sns.pairplot(df, hue='time', diag_kind='hist')
plt.suptitle("Pair Plot of Numerical variables by day", y=1.02, x=0.5)
plt.show()
```

```python
#sns.pairplot(df, kind='scatter')
sns.pairplot(df, hue='time', diag_kind='hist')
plt.suptitle("Pair Plot of Numerical variables by day", y=1.02, x=0.5)
plt.show()
```

#### **9. Heatmap**
- Displaying the correlation matrix of numerical variables.
- Identifying patterns and relationships in a large dataset.

```python
Num = df.select_dtypes(include=['int', 'float']).columns.tolist()
Num
```

```python
correlation_matrix = df[Num].corr()
correlation_matrix
```

```python
sns.heatmap(correlation_matrix, annot=True, cmap='PuBu')
plt.title('Correlation Heatmap')
plt.show()
```

#### **10. Violin Plot**
- Combines the features of box plots and kernel density plots.
- Visualizing the distribution of a numerical variable across different categories.

```python
sns.violinplot(data=df, y='total_bill', split=False, inner_kws=dict(box_width=15, whis_width=2), palette='Set3')
plt.title('Violin Plot of Total Bill by Days', color='blue', fontsize=16, fontweight='bold')
plt.show()
```

```python
sns.violinplot(data=df, x='day', y='total_bill', split=False, inner_kws=dict(box_width=15, whis_width=2), palette='Set3')
plt.title('Violin Plot of Total Bill by Days', color='blue', fontsize=16, fontweight='bold')
plt.show()
```

#### **11. Join Plot**
- Exploring the relationship between two continuous variables.
- Identifying the patterns, correlations, or clusters in the data.

```python
x = 'total_bill'
y = 'tip'
sns.jointplot(x=x, y=y, kind='scatter', data=df, color='red')    # scatter  | kde | hist | hex | reg | resid 
plt.show()
```

#### **12. Swarm Plot**
- Draws a categorical scatterplot with points adjusted to be non-overlapping.
- This function is similar to stripplot(), but the points are adjusted (only along the categorical axis) so that they don't overlap.

```python
sns.set_style('whitegrid')
# Create a swarmplot
sns.swarmplot(x='day', y='total_bill', hue='size', data=df, size=10, legend=True, marker='*')
plt.legend(loc='upper left')
plt.show()
```

#### **13. Strip Plot**
- Draw a categorical scatterplot using jitter to reduce overplotting.
- A Strip plot can be drawn on its own, but it is also a good complement to a box or violin plot in cases where you want to show all observations along with some representation of the underlying distribution.

```python
sns.set(font_scale=1)
sns.stripplot(data=df, x='day', y='total_bill', hue='smoker', size=10)
plt.show()
```

#### **14. SubPlots**

```python
plt.figure(figsize=(12, 8))

plt.subplot(2,2,1)
sns.barplot(data=df, x='sex', y='tip', hue='day', color='lightgreen', errorbar=None, legend=False)
plt.title('Line Plot of Smoker by Tips')

plt.subplot(2,2,2)
sns.scatterplot(data=df, x='size', y='total_bill', color='green', edgecolor='black', marker='o')
plt.title('Bar Plot of Total bill by size')

plt.subplot(2,2,3)
sns.boxplot(data=df, y='tip', color='green')
plt.title('Box Plot of Tip')

plt.subplot(2,2,4)
sns.histplot(data=df, x='size', color='green')
plt.title('Hist Plot of Size')

plt.tight_layout()
plt.show()
```
